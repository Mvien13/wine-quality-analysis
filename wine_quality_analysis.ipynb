```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay

# Load the dataset
df = pd.read_csv('/content/winequality-red.csv')

# 1. Understand the dataset structure
print("Shape of the dataset:", df.shape)
print("\nColumn names and data types:")
print(df.info())
print("\nMissing values in each column:")
print(df.isnull().sum())
print("\nSummary statistics:")
print(df.describe())
print("-------")
print("Result Analyse of the dataset structure ")
print("-------")
print("The dataset contains 1599 rows (wine samples) and 12 columns (11 features + 1 target variable quality)")
print("There are no missing values, which means the dataset is clean and ready for analysis.")
print("alcohol content ranges from 8.4% to 14.9%, with an average of 10.42%.")
print("-------")

# 2. Target variable analysis
plt.figure(figsize=(8, 6))
sns.histplot(df['quality'], bins=10, kde=True, color='blue')
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.show()
print("\nSummary statistics for 'quality':")
print(df['quality'].describe())

print("-------------------------------------")
print("Result Analysis of the qualtiy")
print("-------------------------------------")
print("Most wines are rated 5 or 6 (average quality). Few wines are rated 3, 4, 7, or 8 (low or high quality).")
print("Mean quality: 5.63. Median quality: 6")
print("The majority of wines are of average quality. To stand out in the market, the company should focus on improving quality to reach higher ratings (7 or 8).")
print("--------------------------------------")
# 3. Feature analysis
df.hist(bins=20, figsize=(15, 12), color='green')
plt.suptitle('Histograms of Numerical Features')
plt.show()
plt.figure(figsize=(15, 10))
df.boxplot(figsize=(12, 8), color='blue')
plt.title('Boxplots of Numerical Features')
plt.show()
print("-------------------------------------")
print("Result Analysis of the Feature")
print("-------------------------------------")
print(" In Histograms: Features like residual sugar, chlorides, and free sulfur dioxide are right-skewed, indicating outliers. pH and density are relatively normally distributed.")
print("In Boxplots: Outliers are present in residual sugar, chlorides, and free sulfur dioxide")
print("Outliers in residual sugar and chlorides may indicate measurement errors or unusual wine samples. Investigate these outliers to ensure consistency in production.")
# 4. Correlation analysis
corr_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
print(corr_matrix['quality'].sort_values(ascending=False))
plt.title('Correlation Matrix')
plt.show()
print("\nCorrelation with 'quality':")
print("-------------------------------------")
print("Result Analysis of the Correction")
print("-------------------------------------")
print("alcohol has the highest positive correlation with quality (0.48).volatile acidity has the highest negative correlation with quality (-0.39). Sulphates also positively correlate with quality (0.25).")
print("By the result, ncrease alcohol content and sulphates to improve wine quality.Reduce volatile acidity (which causes unpleasant odors) to enhance quality.")
# 5. Feature-target relationships
plt.figure(figsize=(15, 10))
for i, col in enumerate(df.columns[:-1]):
    plt.subplot(3, 4, i+1)
    sns.scatterplot(x=df[col], y=df['quality'], alpha=0.6, color='purple')
    sns.boxplot(x=df['quality'], y=df[col], palette='viridis')
    plt.title(f'{col} vs Quality')
plt.tight_layout()
plt.show()
print("-------------------------------------")
print("Result Analysis of the Feature-target relationships ")
print("-------------------------------------")
print("As for Scatterplots: Higher alcohol and sulphates correlate with better quality. Higher volatile acidity and chlorides correlate with lower quality.")
print("As for Boxplots: High-quality wines have higher alcohol and sulphates. Low-quality wines have higher chlorides and volatile acidity.")
print ("So, Focus on increasing alcohol and sulphates while reducing volatile acidity and chlorides to improve wine quality.")
# 6. Outlier Detection and Handling
# Cap and floor values to handle outliers
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

plt.figure(figsize=(12, 8))
df.boxplot()
plt.title('Boxplots After Outlier Handling')
plt.xticks(rotation=45)
plt.show()


# 7. Feature engineering insights
df['alcohol_level'] = pd.cut(df['alcohol'], bins=[0, 10, 12, 15], labels=['Low', 'Medium', 'High'])
plt.figure(figsize=(8, 6))
sns.countplot(x='alcohol_level', hue='quality', data=df, palette='viridis')
plt.title('Alcohol Level vs Quality')
plt.show()
print("-------------------------------------")
print("Result Analysis of the Feature engineering insights")
print("-------------------------------------")
print("Alcohol Level: Wines with high alcohol content are more likely to have higher quality scores.")
print(" By aiming to produce wines with higher alcohol content (12% or more) to achieve better quality ratings.")

# Ensure 'quality' is integer type
df['quality'] = df['quality'].astype(int)

# Check unique values in 'quality'
print("Unique values in 'quality':", df['quality'].unique())

# Prepare data for modeling
features = df.drop(columns=['quality', 'alcohol_level'], errors='ignore')  # Exclude target and categorical columns
labels = df['quality']

# Scale features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(features_scaled, labels, test_size=0.2, random_state=42)

# Modeling
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Testing
y_pred = model.predict(X_test)
print(f"Model Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))

#Visualization: Confusion Matrix
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, cmap='viridis')
plt.title("Confusion Matrix")
plt.show()

#Visualization: Feature Importances
feature_importances = pd.Series(model.feature_importances_, index=features.columns)
feature_importances.sort_values(ascending=False).plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Feature Importances from Random Forest Model')
plt.ylabel('Importance')
plt.show()

#Visualization: Predicted vs Actual Quality
plt.figure(figsize=(8, 6))
sns.histplot(y_test, color='blue', label='Actual', kde=True)
sns.histplot(y_pred, color='red', label='Predicted', kde=True)
plt.title('Predicted vs Actual Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Predicting new data
new_data = np.array([[7.4, 0.7, 0.0, 1.9, 0.076, 11.0, 34.0, 0.9978, 3.51, 0.56, 9.4]])
new_data_scaled = scaler.transform(new_data)
predicted_quality = model.predict(new_data_scaled)
print(f"Predicted wine quality for the new data: {predicted_quality[0]}")
```
